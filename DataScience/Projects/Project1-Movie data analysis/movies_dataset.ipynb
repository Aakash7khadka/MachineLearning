 ## Best and worst movies

```python
df_best=df[["title","release_date","budget","revenue","production_companies",'average_rating',"popularity","vote_count"]].copy()
```

```python
df_best.head()
```

```python
df_best['profit']=df_best['revenue']-df_best['budget']
df_best['return']=df_best['revenue'].div(df_best.budget)
```

```python
df_best.head()
```

```python
df_best.set_index('title',inplace=True)
```

```python
df_best.sort_values(by="average_rating",ascending=False)
```

```python
df_best.budget.fillna(0,inplace=True)
df_best.average_rating.fillna(0,inplace=True)
```

```python
df_best.head()
```

```python
df_best[df_best.budget>=10].sort_values(by="return",ascending=False)
```

```python
def best_worst(by,n=5,ascending=False,min_budget=0,min_votes=0):
    return df_best.loc[(df_best.budget>=min_budget) & (df_best.vote_count>=min_votes),[by]].sort_values(by=by,ascending=ascending).head(n).copy()
```

Top 5 movies according to various fields

```python
#by highest revenue
best_worst('revenue')
```

```python
#by highest return on investment
best_worst('return')
```

```python
#by highest profit
best_worst("profit")
```

```python
#by worst profit
best_worst("profit",ascending=True,min_budget=5)
```

```python
# by highest rating
best_worst("average_rating",min_votes=5000)
```

```python
#by worst rating
best_worst("average_rating",min_votes=5000)
```

```python
# most popular
best_worst("popularity")
```

## Movies recommendation

### Action or thriller english movies with rating greater than 7.5

```python
df.genres
```

```python
df.genres.str.contains("Action").sum()
```

```python
df.genres.str.contains("Thriller").sum()
```

```python
action_mask=df.genres.str.contains("Action")
thriller_mask=df.genres.str.contains("Thriller")
avg_rating_mask=df.average_rating>=7.5
voting_count_mask=df.vote_count>5000
```

```python
df.loc[(action_mask | thriller_mask) & avg_rating_mask & voting_count_mask,["title","average_rating"]].sort_values(by="average_rating",ascending=False).head(20)
```
