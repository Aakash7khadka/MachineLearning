```python
#https://www.kaggle.com/c/home-data-for-ml-course/data?select=test.csv

import pandas as pd
train=pd.read_csv('Downloads/train.csv',index_col='Id')
test=pd.read_csv('Downloads/test.csv',index_col='Id')
```

```python
train.head()
```

```python
test.head()
```

```python
#Separate the target variable from training data 
train.dropna(axis=0,subset=['SalePrice'],inplace=True)
y=train.SalePrice
train.drop('SalePrice',axis=1,inplace=True)
```

```python
#Separate the training and validation data
from sklearn.model_selection import train_test_split
X_train,X_valid,y_train,y_valid=train_test_split(train,y,test_size=0.3,random_state=0)
```

```python
#Separate the numerical and categorical columns in training data 
num_col=[col for col in X_train.columns if X_train[col].dtype in(['int64','float64'])]
cat_col=[col for col in X_train.columns if X_train[col].dtype == 'object' and train[col].nunique()<10]
```

```python
#keep the selected columns only
my_cols=num_col+cat_col
X_train=X_train[my_cols].copy()
X_valid=X_valid[my_cols].copy()
test=test[my_cols].copy()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

#define numerical transformation with imputer
num_transform=SimpleImputer(strategy='median')

#define categorical transformation with imputation and onehotencoding
cat_transform=Pipeline(steps=[
    ('impute',SimpleImputer(strategy='most_frequent')),
    ('onehot',OneHotEncoder(handle_unknown='ignore'))
])

#column transformation completes the preprocessing 
preprocessing=ColumnTransformer(transformers=[
    ('num',num_transform,num_col),
    ('cat',cat_transform,cat_col)
])

#define a model
model=RandomForestRegressor(n_estimators=100,random_state=0)

clf=Pipeline(steps=[
    ('pre',preprocessing),
    ('model',model)
])


#fit the training data to model
clf.fit(X_train,y_train)

#predict 
pred=clf.predict(X_valid)

#Calculate the error
print("Mean absolute error:{}".format(mean_absolute_error(y_valid,pred)))
```

```python
#prediction for the test data
clf.predict(test)
```
